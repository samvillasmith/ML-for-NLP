# Bag of Words (BoW)
The Bag of Words model is a way of representing text data when modeling text with machine learning algorithms. In this model, a text (like a sentence or a document) is represented as a multiset (bag) of its words, disregarding grammar and even the order of words but keeping multiplicity.

The process typically involves:

Tokenization: Splitting the text into individual words or tokens.
Vocabulary Creation: Creating a list of all unique words in the entire dataset.
Vectorization: Representing each text document as a vector where each dimension corresponds to a word in the vocabulary. The value in each dimension is typically the frequency of the word in the document or a weighted frequency like TF-IDF.
Essentially, it counts how many times each word appears in a document.

```python
import pandas as pd

path = "/content/smsspamcollection.csv"

# CSV with header, remove BOM, and rename columns
messages = pd.read_csv(
    path,
    sep=",",
    header=0,              # file has a header row
    encoding="utf-8-sig",  # strips the BOM
    names=["label", "message"]
)

messages.head()
```

```python
# Data cleaning and preproccessing
import re
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
  review = re.sub('[^a-zA-z]', ' ', messages['message'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
# Create the BoW Model
from sklearn.feature_extraction.text import CountVectorizer
# Enable binary option for BoW
cv = CountVectorizer(max_features=100, binary=True)
```

```python
X = cv.fit_transform(corpus).toarray()
```

## N-grams

N-grams are contiguous sequences of N items from a given sample of text or speech. The items can be phonemes, syllables, letters, words or base pairs according to the application. N-grams are typically collected from a text or speech corpus.

-   **Unigrams (N=1):** Single words.
-   **Bigrams (N=2):** Pairs of consecutive words.
-   **Trigrams (N=3):** Sequences of three consecutive words.
-   And so on for higher values of N.

N-grams are used in various natural language processing tasks, including:

-   **Language modeling:** Predicting the next word in a sequence.
-   **Text classification:** Using the frequency of N-grams as features.
-   **Spell checking and correction:** Identifying unlikely N-grams.
-   **Information retrieval:** Matching N-grams in queries and documents.

They capture some of the local context and word order, which is lost in the Bag of Words model.
